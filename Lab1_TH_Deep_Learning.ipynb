# **Bài 20**

```
import numpy as np

# Tạo 2 Series ngẫu nhiên 20 phần tử, giá trị từ 1 đến 50
x = pd.Series(np.random.randint(1, 51, size=20))
y = pd.Series(np.random.randint(1, 51, size=20))

# Hội (Union): Lấy tất cả các phần tử của cả x và y (loại bỏ các giá trị trùng lặp)
union_result = pd.Series(np.union1d(x, y))
print("Hội (Union) của x và y:\n", union_result)

# Giao (Intersection): Lấy các phần tử chung xuất hiện ở cả x và y
intersect_result = pd.Series(np.intersect1d(x, y))
print("\nGiao (Intersection) của x và y:\n", intersect_result)
```

# **Bài 15 - 16**

```
class GeometryModel(tf.Module):
    def __init__(self, name=None):
        super().__init__(name=name)

    # Bài 15: P = (a + b) * 2
    def calc_perimeter(self, a, b):
        return tf.multiply(tf.add(a, b), 2.0)

    # Bài 16: S_xq = (a + b) * c * 2 | S_tp = (a * b) * 2 + S_xq
    def calc_surface_area(self, a, b, c):
        s_xq = tf.multiply(tf.multiply(tf.add(a, b), c), 2.0)
        s_day = tf.multiply(tf.multiply(a, b), 2.0)
        s_tp = tf.add(s_day, s_xq)
        return s_tp
```

# **Bài 17 - 18- 19**

```
class GeometryLogicProofModel(tf.Module):
    def __init__(self, name=None):
        super().__init__(name=name)

    # Bài 17: Kiểm tra cạnh bên hình thang song song (AD//BC <=> AD=BC và AB=CD)
    def prove_parallel_sides(self, AD, BC, AB, CD):
        cond1 = tf.equal(AD, BC)
        cond2 = tf.equal(AB, CD)
        return tf.logical_and(cond1, cond2)

    # Bài 18: Chứng minh hai cạnh đáy hình thang bằng nhau (AB=CD <=> AD=BC và AD//BC)
    def prove_equal_bases(self, AD, BC, is_AD_parallel_BC):
        cond1 = tf.equal(AD, BC)
        return tf.logical_and(cond1, is_AD_parallel_BC) # is_AD_parallel_BC là boolean

    # Bài 19: Tính chất đường trung bình tam giác (NA = NC)
    def prove_midpoint_N(self, AM, MB, is_MN_parallel_BC):
        # Nếu M là trung điểm AB (AM = MB) và MN // BC thì N sẽ là trung điểm AC (NA = NC)
        is_M_midpoint = tf.equal(AM, MB)
        return tf.logical_and(is_M_midpoint, is_MN_parallel_BC)
```
